Installing dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Load csv file to pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#First 5 rows
credit_card_data.head()
```

```python
#Last 5 rows
credit_card_data.tail()
```

```python
#Dataset information
credit_card_data.info()
```

```python
#Checking number of missing values in each column

credit_card_data.isnull().sum()
```

```python
#Distribution of legit transaction and fraudlent transaction

credit_card_data['Class'].value_counts()
```

```python
#seperating data for analysis

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
#Stastical measures of data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Comparing the values of both the transactions

credit_card_data.groupby('Class').mean()
```

Dealing with the Imbalanced Data

Undersampling :
We got number of fraudulent data = 173

```python
legit_sample = legit.sample(n=173)
```

```python
#Concatinate two data frames

new_dataset = pd.concat([legit_sample,fraud],axis = 0)
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
#Splitting the data into features and targets

x = new_dataset.drop(columns='Class',axis = 1)
y = new_dataset['Class']

#Splitting the data into training data and testing data

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, stratify = y, random_state = 2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model Training

model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

Model Evaluation


Accuracy Score


```python
#Accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print(training_data_accuracy)
```

```python
#Accuracy on testing data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
print(testing_data_accuracy)
```
